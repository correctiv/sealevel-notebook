```python
#get the data from worldbank, Stand: 30. Oktober 2017

emissions = pd.read_csv('context_data/c02_emissions_worldbank_2014.csv', usecols=[0, 1, 48, 58])

#emissions['c02_rank'] = emissions['2013'].rank(ascending=False)
emissions.rename(columns={'Country Code': 'country_code'}, inplace=True)
emissions.rename(columns={'Country Name': 'country_name'}, inplace=True)
#emissions.rename(columns={'2014': 'co2_emissions'}, inplace=True)

#emissions.sort_values(by='c02_rank').head()

emissions.head()
```

```python
emissions_europe = emissions.merge(country_list, left_on='country_code', right_on='ISO-3', how='inner')
```

```python
emissions_europe
```

```python
worldbank_meta = pd.read_csv('../../../Downloads/API_EN.ATM.CO2E.PC_DS2_en_csv_v2/Metadata_Country_API_EN.ATM.CO2E.PC_DS2_en_csv_v2.csv')
```

```python
worldbank_meta = worldbank_meta[['Country Code', 'IncomeGroup']]
```

```python
worldbank_meta
```

```python
emissions_europe_meta = emissions_europe.merge(worldbank_meta, left_on='country_code', right_on='Country Code', how='inner')
```

```python
emissions_europe_meta['change'] = (emissions_europe_meta['2014'] - emissions_europe_meta['2004']) / emissions_europe_meta['2004']
```

```python
emissions_europe_meta = emissions_europe_meta[['Country', '2004', '2014', 'IncomeGroup', 'change']]
emissions_europe_meta['2004'] = emissions_europe_meta['2004'].round(decimals=2)

emissions_europe_meta['2014'] = emissions_europe_meta['2014'].round(decimals=2)

emissions_europe_meta['change'] = emissions_europe_meta['change'].round(decimals=2)
```

```python
emissions_europe_meta.to_csv('../../../Desktop/co2_emissions_europe.csv', index=False)
```
